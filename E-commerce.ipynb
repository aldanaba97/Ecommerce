##2) TRANFORMACION DE DATOS

Se elaborara diferentes funciones que transformen ciertos datos de los dataframe evitando problemas para futuras operaciones. Se tomara decisiones en cuanto a la cantidad de nulos y asi, lograr una eficacia en el analisis que se pretenda realizar.

Teniendo en cuenta que los campos de tipo Fecha los reconoce como Object se elaborara un metodo que tranforme los datos al formato ideal, en este caso seria, Datetime.

```python
def TransformarFecha(df, nombre_columna_fecha): #El metodo recibe como parametro el dataframe y la/s columnas que deseen modificar.
   df[nombre_columna_fecha] = df[nombre_columna_fecha].apply(pd.to_datetime) #Se realiza un casteo donde se le indica que la columna pasa a ser datetime
   return df
```

Al igual que el caso anterior, las columnas de tipo String son consideradas como Object. Se realizara una funcion que permita realizar un casteo.


```python
def TransformarString(df, nombre_columna_string):  #El metodo recibe como parametro el dataframe y la/s columnas que deseen modificar.
 for columna in nombre_columna_string:
  df[nombre_columna_string] = df[nombre_columna_string].astype("string"); #Se realiza un casteo donde se le indica que la columna pasa a ser string
  return df
```

En la siguiente funcion se transformara los registros en minuscula para evitar inconvenientes con analisis de filtrado, etc.

```python
 def TransformacionMinuscula(df, columna): #El metodo recibe como parametro el dataframe y la/s columnas que deseen modificar.
  df[columna] = df[columna].str.lower(); #Se realiza una modificacion donde todos los registros deben estar en minuscula para la columna ingresada.
  return df
```

##ANALISIS DE DATOS
Se realizaran diferentes analisis teniendo en cuenta la problematica descripta anteriormente, elaborando insighs que ayuden al progreso comercial de la empresa. Se comenzara con consultas generales que puedan ayudar a otras otras operaciones mas complejas.

```python
#Esta funcion va a brindar informacion general de cada columna
def analisisDF(df):
  print("------------------------------------------------")
  print("Variables cuantitativas: ")
  columnas_numericas= df.select_dtypes(include=['int', 'float']).columns; #El select_dtypes devuelve el tipo de columna que se haya especificado, en este caso las columnas de tipo int y float.
  if not columnas_numericas.empty :  #Realizamos un manejo de error con la condicion de que solo puedan ser descriptas las columnas que no esten vacias
    print(df[columnas_numericas].describe()); #Se muestra la descripcion de cada columna
    print("------------------------------------------------")
    print("Variables cualitativas")
  columnas_string = df.select_dtypes(include=['string']).columns; #En este caso el select_dtypes devuelve columnas de formato string
  if not columnas_string.empty:
    print(df[columnas_string].describe());
```
